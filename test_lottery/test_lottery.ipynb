```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_ticks_evolution_from_csv(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Convert relevant columns to integers
    df['step'] = df['step'].astype(int)
    df['tickets'] = df['tickets'].astype(int)
    df['ticks'] = df['ticks'].astype(int)

    # Plotting
    fig, ax = plt.subplots()
    processes = df['pid'].unique()
    colors = plt.cm.jet(range(0, 256, int(256/len(processes))))

    for i, pid in enumerate(processes):
        process_data = df[df['pid'] == pid]
        ax.plot(process_data['step'], process_data['ticks'], label=f'PID {pid}, Tickets {process_data["tickets"].iloc[0]}', color=colors[i])

    ax.set_xlabel('Step')
    ax.set_ylabel('Ticks')
    ax.set_title('Evolution of Ticks for Each Process')
    ax.legend()
    plt.show()

# Example usage
file_path = 'test_output.csv'  # Path to your CSV file
plot_ticks_evolution_from_csv(file_path)
```
